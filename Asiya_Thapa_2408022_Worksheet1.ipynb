Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A^-1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)^T = B^TA^T

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1 = I:")
print(I)

AB = np.dot(A, B)
BA = np.dot(B, A)
AB_not_equal_BA = not np.array_equal(AB, BA)
print("\nAB != BA:", AB_not_equal_BA)

AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
AB_T_equals_BT_AT = np.array_equal(AB_T, BT_AT)
print("\n(AB)^T = B^T * A^T:", AB_T_equals_BT_AT)
```

4.2 Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.

```python
import time

# Create two Python lists with size 1,000,000
list_size = 1_000_000
list1 = list(range(list_size))
list2 = list(range(list_size))

# Measure the time taken for element-wise addition using Python lists
start_time = time.time()
list_result = [x + y for x, y in zip(list1, list2)]
end_time = time.time()

# Calculate and print the elapsed time
elapsed_time = end_time - start_time
print(f"Time taken for element-wise addition using Python lists: {elapsed_time:.6f} seconds")




```

Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import numpy as np
import time

# Create two NumPy arrays with size 1,000,000
array_size = 1_000_000
array1 = np.arange(array_size)
array2 = np.arange(array_size)

# Measure the time taken for element-wise addition using NumPy arrays
start_time = time.time()
array_result = array1 + array2
end_time = time.time()

# Calculate and print the elapsed time
elapsed_time = end_time - start_time
print(f"Time taken for element-wise addition using NumPy arrays: {elapsed_time:.6f} seconds")
```

Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.

```python
import time

# Create two Python lists with size 1,000,000
list_size = 1_000_000
list1 = list(range(list_size))
list2 = list(range(list_size))

# Measure the time taken for dot product using Python lists
start_time = time.time()
dot_product = sum(x * y for x, y in zip(list1, list2))
end_time = time.time()

# Calculate and print the elapsed time
elapsed_time = end_time - start_time
print(f"Dot product using Python lists: {dot_product}")
print(f"Time taken for dot product using Python lists: {elapsed_time:.6f} seconds")
```

sure and print the time taken for this operation.

```python
import numpy as np
import time

# Create two 1000x1000 matrices using NumPy arrays
matrix_size = 1000
matrix1 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
matrix2 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)

# Measure the time taken for matrix multiplication using NumPy arrays
start_time = time.time()
result = np.dot(matrix1, matrix2)  # Or equivalently, matrix1 @ matrix2
end_time = time.time()

# Calculate and print the elapsed time
elapsed_time = end_time - start_time
print(f"Time taken for matrix multiplication using NumPy arrays: {elapsed_time:.6f} seconds")
```
